```python
import pandas as pd
```

```python
data_all = "Output/2019_2020_mixed_beverage"


df_all = pd.read_csv(data_all, encoding="utf-8")
```

```python
#efficient way to extract year from string format date
df_all['year'] = pd.DatetimeIndex(df_all['obligation_end_date_yyyymmdd']).year
df_all['month'] = pd.DatetimeIndex(df_all['obligation_end_date_yyyymmdd']).month
df_all.head()
```

```python
mask_2019 = df_all.year == 2019
df_2019 = df_all[mask_2019]
location_sales_2019 = pd.DataFrame(df_2019.groupby('location_address')['total_receipts'].sum())
location_sales_2019.head()
```

```python
bins = [0,100000,250000,70000000]
group_names=["low",'medium','high']
location_sales_2019["volume_category"] = pd.cut(location_sales_2019['total_receipts'],bins,labels=group_names,include_lowest=True)
location_sales_2019
```
